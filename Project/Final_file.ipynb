```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python

chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=chrome_options)
driver.maximize_window()
driver.get("https://daraz.pk/")
```

```python
def switch_to_new_window(driver):
    try:
        ActionChains(driver).key_down(Keys.CONTROL).click(element).key_up(Keys.CONTROL).perform()
        new_window_handle = next(handle for handle in driver.window_handles if handle != driver.current_window_handle)
        driver.switch_to.window(new_window_handle)
        return True
    except Exception as e:
        print(f"Error: {e}. Unable to switch to the new window.")
        return False

price=[]
brand=[]
name=[]
prodID=[]
templist=[]
url=[]
score=[]
review=[]
ID=[]
ratings=[]

RAM=[]
ROM=[]
BATTERY=[]
CAMERA=[]

ram=["16","12","8","6","4","3","2"]
rom=["512","256","128","64","32"]
battery=["4000","4600","5000","5050","6000"]
camera=["108","50","48","32","16","13"]

for x in range(7, 12):  # 7,12
    for i in range(1, 41):  # 1,41
        reviewtemp = []
        IDtemp = []
        element_xpath = f"/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[2]/div[{i}]"
        element = driver.find_element(By.XPATH, element_xpath)

        if not switch_to_new_window(driver):
            # Skip the rest of the loop if switching to the new window fails
            continue

        try:
            # price
            price_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "./html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[7]/div/div/span")))
            price.append(price_element.text)

            # name
            name_element = driver.find_element(By.XPATH, "./html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[3]/div/div/span")
            name.append(name_element.text)

            # brand
            brand_element = driver.find_element(By.XPATH, "./html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[5]/div/a[1]")
            if brand_element.text == "No Brand":
                templist = (name_element.text).split()
                brand.append(templist[0])
            else:
                brand.append(brand_element.text)

            # Product ID
            prodID.append(f"ID NO : {x}{i}")
            url.append(driver.current_url)

            # Reviews
            ello1 = 0
            try:
                noofrev_element = driver.find_element(By.XPATH, "./html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[2]/div[1]/div[1]/div[3]")
                ello = noofrev_element.text
                ratings.append(ello)
                tempo = ello.split()
                ello1 = int(tempo[0])
            except NoSuchElementException:
                pass  # Handle the case where no element is found gracefully

            itera = 0
            if ello1 > 50:
                revbutton = driver.find_element(By.XPATH, "./html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[3]/div[4]/ul/li[10]/div/div")
                revbutton.click()
                itera = 51
                time.sleep(1)
                p50button = driver.find_element(By.XPATH,"./html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[3]/div[4]/ul/li[10]/div/div[2]/div/div/div/div/div/div[4]/div")
                p50button.click()

            time.sleep(1)

            # Reviews
            base_xpath = "/html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[3]/div"
            for j in range(1, itera):
                xpathuser = f"{base_xpath}[{j}]/div[1]/div[2]"
                xpathrev = f"{base_xpath}[{j}]/div[2]"
                review_element = driver.find_element(By.XPATH, xpathrev)
                reviewtemp.append(review_element.text)

                ID_element = driver.find_element(By.XPATH, xpathuser)
                IDtemp.append(ID_element.text)

            if ello1 > 0:
                score_element = driver.find_element(By.XPATH,"./html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[2]/div[1]/div[1]/div[1]/span[1]")
                score.append(score_element.text)
            else:
                score.append("No Score")

            review.append(reviewtemp if reviewtemp else 'No Review')
            ID.append(IDtemp if IDtemp else "No ID")
            
            specbox = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "./html/body/div[4]/div/div[9]/div[1]/div[2]/div/div/div")))
            driver.execute_script("arguments[0].scrollIntoView();", specbox)
            temp=specbox.text
            temp=temp.replace("1080",'').replace("Android 13",'').replace("Android13",'')
            #print(temp)
            time.sleep(2)
            print("---------------")
            for j in range(len(ram)):
                if ram[j] in temp:
                    RAM.append(ram[j])
                    print(f"ram :  {ram[j]}")
                    break
                if j==len(ram)-1:
                    RAM.append(None)
    
            for j in range(len(rom)):
                if rom[j] in temp:
                    ROM.append(rom[j])
                    print(f"rom : {rom[j]}")
                    break
                if j==len(rom)-1:
                    ROM.append(None)
          
            for j in range(len(battery)):
                if battery[j] in temp:
                    BATTERY.append(battery[j])
                    print(f"battery : {battery[j]}")
                    break
                if j==len(battery)-1:
                    BATTERY.append(None)
            for j in range(len(camera)):
                if camera[j] in temp:
                    CAMERA.append(camera[j])
                    print(f"camera : {camera[j]}")
                    break
                if j==len(camera)-1:
                    CAMERA.append(None)
            print("---------------")

        except Exception as e:
            print(f"Error: {e}. Skipping iteration.")

        finally:
            driver.close()

            # Switch back to the original window
            original_window_handle = driver.window_handles[0]
            driver.switch_to.window(original_window_handle)

    base_xpath = "/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li"
    if x <= 9:
        xpath = f"{base_xpath}[{9}]/a"
    else:
        xpath = f"{base_xpath}[{x}]/a"

    page = driver.find_element(By.XPATH, xpath)
    page.click()
    time.sleep(2)
```

```python
# price from RS 33,999 -> 33999
price_str=[]
for i in range(len(price)):
    price_str.append(price[i][4:])
    price_str[i]=price_str[i].replace(',','')
price_int = list(map(int, price_str))
```

```python
#making database
import sqlite3
conn = sqlite3.connect('finalDataBaseProject.db')

# Create a cursor
cursor = conn.cursor()

# Execute the CREATE TABLE statements
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Products (
        product_id TEXT,
        name VARCHAR(255),
        price INTEGER,
        brand VARCHAR(255),
        url TEXT
    );
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS Reviews (
        review_id TEXT,
        product_id TEXT,
        review_text TEXT,
        rating TEXT,
        FOREIGN KEY (product_id) REFERENCES Products(product_id)
    );
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS Specs (
        Camera TEXT,
        Ram TEXT,
        Rom TEXT,
        Battery TEXT,
        product_id TEXT,
        FOREIGN KEY (product_id) REFERENCES Products(product_id)
    );
''')
conn.commit()
conn.close()

```

```python
#insertion in database
import sqlite3
conn = sqlite3.connect('finalDataBaseProject.db')
cursor = conn.cursor()

for i in range(len(prodID)):
    cursor.execute('''
        INSERT INTO Products (product_id, name, price, brand, url,rating,noofrev)
        VALUES (?, ?, ?, ?, ?, ?, ?);
    ''', (prodID[i], name[i], price_int[i], brand[i], url[i],score[i],ratings[i]))

# Insert data into the Reviews table
for i in range(len(review)):
    if ID[i] != "No ID":
        for j in range(len(ID[i])):
            cursor.execute('''
                INSERT INTO Reviews (review_id, product_id, review_text)
                VALUES (?, ?, ?);
            ''', (ID[i][j],prodID[i], review[i][j]))
    else:
        cursor.execute('''
                INSERT INTO Reviews (review_id, product_id, review_text)
                VALUES (?, ?, ?);
            ''', (ID[i],prodID[i], review[i]))
        

for i in range(len(prodID)):
        cursor.execute('''
            INSERT INTO Specs (product_id, Camera, Ram, Rom, Battery)
            VALUES (?, ?, ?, ?, ?);
        ''', (prodID[i], CAMERA[i], RAM[i], ROM[i], BATTERY[i]))

        

conn.commit()
conn.close()

```
